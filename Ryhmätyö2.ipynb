```python
import kagglehub
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Download the dataset
path = kagglehub.dataset_download("shreyasur965/democracy-index")

# List the contents of the directory where the dataset is saved
dataset_files = os.listdir(path)

# Check if there are any files to load
if dataset_files:
    # Assuming the dataset is a CSV file and we are using the first file in the list
    csv_path = os.path.join(path, dataset_files[0])  # Replace with a specific file if necessary
    data = pd.read_csv(csv_path)

    # Print the first few rows of the dataset to inspect its structure
    print(data.head())

    # Check if the dataset contains the required columns for 3D plotting
    if 'Year' in data.columns and 'Democracy score' in data.columns and 'Entity' in data.columns:
        # Convert 'Entity' to a numerical value (categorical encoding)
        data['Entity_num'] = data['Entity'].astype('category').cat.codes

        # Create the 3D plot
        fig = plt.figure(figsize=(12, 8))
        ax = fig.add_subplot(111, projection='3d')

        # Scatter plot using Year, Democracy score, and Entity_num
        ax.scatter(data['Year'], data['Democracy score'], data['Entity_num'], c=data['Democracy score'], cmap='viridis', marker='o')

        # Customizing the plot
        ax.set_xlabel('Year')
        ax.set_ylabel('Democracy Score')
        ax.set_zlabel('Entity (Numerical Encoding)')
        ax.set_title('3D Plot of Democracy Score vs Year vs Entity')

        # Show the plot
        plt.show()

    else:
        print("Dataset does not contain 'Year', 'Democracy score', or 'Entity' columns.")

else:
    print("No files found in the dataset directory.")
```
